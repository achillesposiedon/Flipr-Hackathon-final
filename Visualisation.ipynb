```python
import pandas as pd
import numpy as np

trdt=pd.read_excel('/home/aaftab/Github/Flipr-Hackathon/Train_dataset.xlsx')
tstdt=pd.read_excel('/home/aaftab/Github/Flipr-Hackathon/Test_dataset.xlsx')
```

```python
def basicvisualisation(df): 
        #function for basic visualistaion gives number of columns,number of rows,first 5 rows of dataset,number of columns with each type dtypes
       
        s=df
        s.ncol=len(s.columns)
        s.nrows=len(s)
        
        print(s.head(5))
        
        print('Number of columns is '+ str(s.ncol))
        print('Number of rows is '+ str(s.nrows))
        
        print(s.dtypes.value_counts()) #prints number of columns of each datatype
        
        if s.ncol<10: # print all column names only if number of columns is less than 10, to keep it neat
            print('Column names are '+s.columns)
         #unique ids in categorical column
        categoricalcolumns=s.select_dtypes(include=['object']).columns
        for i in categoricalcolumns:
            print('The categorical column '+str(i)+' has '+str(len(s[str(i)].unique()))+' unique ids ')

basicvisualisation(trdt)
    
```

```python
basicvisualisation(tstdt)
```

```python
def columnwisedata(dft): #gives column wise data like datatype,number of null values,percentage of null values,max,min,mean,count.
    
        s=dft
        df = {}
        
        df['dtypes'] = s.dtypes
        
        df['null_sum'] = s.isnull().sum()
        df['null_per'] = s.isnull().mean()*100
        
        df['max'] = s.max()
        df['min'] = s.min()
        
        df['count'] = s.count()
        df['mean'] = s.mean()
    
        
        if len(s.columns)>20: #return dataframe if number of columns is greater than 20
            return pd.DataFrame(df)
        else :
            print(pd.DataFrame(df)) #directly print output if number of columns is less than 20
    
```

```python
columnwisedata(trdt)
```

```python
columnwisedata(tstdt)
```

```python
def correlation(df):
        s=df
        corrmatrix=s.corr()
        return (corrmatrix.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2))
        
```

```python
correlation(trdt)
```

```python
correlation(tstdt)
```

```python
def onehotencode(df,colids):
        #colids names as a list
        df = pd.get_dummies(df, columns=colids)
        return df
def fillmeanvalue(df):
        df.fillna(df.mean(), inplace=True)
        return(df)
```

```python
trobjmean=onehotencode(trdt,['Designation','Gender','Married','Mode_transport','Occupation','Pulmonary score','Region','cardiological pressure','comorbidity'])
trobjmean=fillmeanvalue(trdt)
#trobjmean.to_csv('Mean-Filled.csv')
```

```python
tstobjmean=onehotencode(tstdt,['Designation','Gender','Married','Mode_transport','Occupation','Pulmonary score','Region','cardiological pressure','comorbidity'])
tstobjmean=fillmeanvalue(tstdt)
#tstobjmean.to_csv('Mean-Filled-test.csv')
```
